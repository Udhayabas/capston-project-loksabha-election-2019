```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
pio.renderers.default="svg"
```

# Q1)which are the top 3 states with most parlimentary constituencies             

```python
Num_cons = data.groupby('STATE')['CONSTITUENCY'].nunique().sort_values(ascending=False).reset_index()
```

```python
Num_cons.head(5)
```

# Q2)which states has the least number of parlimentary constituencies

```python
Num_const = data.groupby('STATE')['CONSTITUENCY'].nunique().sort_values(ascending=True).reset_index()
Num_const.head(5)
```

```python
ax = px.bar(Num_cons,y='CONSTITUENCY',x='STATE',color = 'CONSTITUENCY', title='The Number of Constituencies from each State')
ax.show()
```

```python
Num_cons
```

# Q3)Give some statics about Education level of candidates who contested in election

```python
education = data['EDUCATION'].value_counts().reset_index()
```

```python
education
```

```python
education.columns = ['EDUCATION','COUNT']
ax = px.bar(education,x = 'EDUCATION', y = 'COUNT',color = 'EDUCATION', title= 'Education Level of the Candidates')
ax.show()
```

# Q4)give winning candidates education level

```python

winner = data[data['WINNER']==1]
educationlevel=winner['EDUCATION'].value_counts().reset_index()
educationlevel.columns=['EDUCATION','COUNT']
educationlevel
```

```python
ax = px.bar(winner,x = 'EDUCATION',y = 'WINNER', title='Winning Candidates Educational Degree').update_xaxes(categoryorder = "total descending")
ax.show()
```

# q5)details of youngest cndidates who won election

```python
young_winner = data[data['WINNER']==1]
young_winner = young_winner.sort_values('AGE')
                                                
```

```python
young_winner.head(3)
```

```python
young_winner = young_winner.sort_values('AGE').head(10)
ax = px.bar(young_winner,x = 'NAME',y = 'AGE',color = 'AGE',hover_data = ['PARTY','STATE','CONSTITUENCY'], title='Youngest Winners')
ax.show()
```

# q6)winning candidates profile of oldest

```python
old_winner = data[data['WINNER']==1]
old_winner = old_winner.sort_values('AGE',ascending = False).head(3)
old_winner
```

```python
ax = px.bar(old_winner,x = 'NAME',y = 'AGE',color = 'AGE',hover_data = ['PARTY','STATE','CONSTITUENCY'], title = 'Oldest Winners and their Details:')
ax.show()
```

# q7)which party has the highst number of criminal cases among winning candidates

```python
criminal_cases = data.groupby('PARTY')['CRIMINALCASES'].sum().reset_index().sort_values('CRIMINALCASES',ascending=False).head(5)
criminal_cases
```

```python
ax = px.bar(criminal_cases, x = 'PARTY',y = 'CRIMINALCASES',color = 'PARTY', title='Total Criminal Cases by respective parties')
ax.show()
```

# q8)highest number of female winner from whichstate

```python
Female_winners = data[(data['WINNER']==1) & (data['GENDER']=='FEMALE')]
Female_winners.head(5)
```

```python
ax = px.histogram(Female_winners, 'STATE', title = 'Female Winners from different States')
ax.show()
```

highest number of female winner from westbangal

# q9) highest number of male winner from which state

```python
male_winners = data[(data['WINNER']==1) & (data['GENDER']=='MALE')]
male_winners.head(5)
```

```python
ax = px.histogram(male_winners, 'STATE', title='Male Winners from different States')
ax.show()
```

# q10)which are the top ten states with most votes

```python
votes = data.groupby('STATE')['TOTAL VOTES'].sum().sort_values(ascending = False).reset_index()
votes.head(10)
```

```python
ax = px.bar(votes,x = 'STATE',y = 'TOTAL VOTES',color='STATE', title='Total Votes cast State Wise')
ax.show()
```

# Q11)which catagory of candidates won more seats

```python
df = data[data['WINNER']==1]
category = df['CATEGORY'].value_counts().reset_index()
category.columns= ['CATEGORY','COUNT']
category
```

```python
ax = px.bar(category,x = 'CATEGORY', y = 'COUNT', color = 'CATEGORY', title='Winners from Various Categories')
ax.show()
```
