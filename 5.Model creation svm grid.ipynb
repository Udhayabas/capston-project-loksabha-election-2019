```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
dataset=pd.read_csv('preprocessedloksabha.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['ASSETS', 'LIABILITIES',
       'GENERAL VOTES', 'POSTAL VOTES','TOTAL VOTES' ]]
```

```python
indep
```

```python
dep=dataset[['WINNER']]
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,100,1000,2000,3000]} 



grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid.fit(x_train, y_train) 
 
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(x_test)


from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)


from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
grid_predictions
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print(cm)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score

# Assuming best_model is your trained classifier and supports predict_proba
best_model = grid.best_estimator_  # Replace this with your actual trained model

# Assuming x_test is your test data
y_pred_prob = best_model.decision_function(x_test)  # Get decision function scores

# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, y_pred_prob)
print("ROC AUC Score:", roc_auc)





```
